```python
from dist_metrics import weighted_euclidean
from dist_metrics import pairwise_dist_wrapper
from dist_metrics import all_pairwise_dist
from dist_metrics import sum_grouped_dist

try:
    from dist_metrics import squared_sum_grouped_dist
except:
    def squared_sum_grouped_dist(pair_list, data, w):
        """
        Return the sum of squared distance
        """
        dist = all_pairwise_dist(pair_list, data, w)
        dist_squared = [ i * i for i in dist]
        return sum(dist_squared)
```

```python
sample_data = pd.read_csv(DATA_PATH + '/sample_2_classes_100.data', header = False)
sample_data.head(3)
```

```python
from itertools import combinations
from random import shuffle

idx_ones = [i for i, val in enumerate(sample_data.y) if val == 1]
idx_zeros = [i for i, val in enumerate(sample_data.y) if val == 0]

sim_pairs, diff_pairs = [], []

for i in combinations(idx_ones, 2):
    sim_pairs.append(i)
    
for i in combinations(idx_zeros, 2):
    sim_pairs.append(i)
    
for a in idx_ones:
    for b in idx_zeros:
        diff_pairs.append((a, b))
        
shuffle(diff_pairs)
```

```python
squared_sum_grouped_dist(sim_pairs, sample_data, [1, 1, 1])
```

```python
sum_grouped_dist(diff_pairs, sample_data, [1, 1, 1])
```

```python
G = lambda w: squared_sum_grouped_dist(sim_pairs, sample_data, w) - np.log(sum_grouped_dist(diff_pairs, sample_data, w))
```

```python
mini_obj_func = lambda w: squared_sum_grouped_dist(sim_pairs, sample_data, w)
const_func = lambda w: sum_grouped_dist(diff_pairs, sample_data, w)
```

```python
from scipy.optimize import minimize
import time

bnds = ((0, None), (0, None), (0, None))
G = lambda w: squared_sum_grouped_dist(sim_pairs, sample_data, w) / squared_sum_grouped_dist(diff_pairs, sample_data, w)

start_time = time.time()
res = minimize(G, (1, 1, 1), method="L-BFGS-B", bounds = bnds)
duration = time.time() - start_time
print("--- %s seconds ---" % duration)
```
